```python
from sklearn.neighbors import KernelDensity
import numpy
from numpy import nan

def get_scaled_density(my_kde, test_input, X, y):
    test_density = my_kde.score_samples(test_input)
    min_density = np.nanmin(test_density[test_density != -np.inf])
    test_density = np.where(test_density == -np.inf, min_density, test_density)
    test_density = [element - min_density for element in test_density]
    max_density = max(test_density)
    my_const=max(y)/max_density
    test_density = [element * my_const for element in test_density]
    return test_density

test_input = numpy.arange(0, max(X), 50).reshape(-1, 1)
kde = KernelDensity(bandwidth=1000, kernel='tophat').fit(X_train)
scaled_density = get_scaled_density(kde, test_input, X_train, y_train)
plt.figure(figsize=(21,13))
plt.plot(test_input,scaled_density, color='red', label='scaled kernel density')
plt.scatter(X_train, y_train, label='train data')
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
# # my_const=int(max(y_train)/max_density)
# my_const = 1000000

def my_kernel(distances):
    my_lambda = 25
    results = []
    for distance in distances:
        result = kde.score_samples(distance.reshape(-1, 1))
        results.append(result)
#     weights = kde.score_samples(distances) # Compute an array of weights however you want
    return results

KNR = KNeighborsRegressor(weights=my_kernel)
KNR.fit(X_train, y_train)
knr_test_predicted = KNR.predict(X_test)
print(mean_squared_error(knr_test_predicted, y_test))
```
